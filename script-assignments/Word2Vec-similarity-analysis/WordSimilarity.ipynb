```python
import random as rand
import string

import gensim.downloader
import nltk
```

Loading questions-words.txt corpus into a set of words

```python
text = open("sherlock_holmes.txt", "r", encoding="utf-8").read()
text = text.replace("\n", " ")
text = text.translate(str.maketrans('', '', string.punctuation))

nltk.download('punkt')
data = nltk.tokenize.word_tokenize(text)
data = list(set(data))
```

```python
print(len(data))
for d in data:
    print(d)
```

Loading pre-trained word2vec model

```python
wv = gensim.downloader.load('word2vec-google-news-300')
```

Cosine similarity function

```python
def similarity(word1, word2):
    return wv.similarity(word1, word2)
```

Similarity analysis on randomly selected pairs of words from the corpus

```python
pairs = []
while len(pairs) < 10:
    word1, word2 = rand.sample(data, 2)
    if (word1, word2) not in pairs and (word2, word1) not in pairs:
        pairs.append((word1, word2))
for pair in pairs:
    try:
        print('%r\t%r\t%.2f' % (pair[0], pair[1], similarity(pair[0], pair[1])))
    except KeyError as err:
        print(err)
```

Function for getting n most similar words of a word

```python
def most_similar(word, n):
    return wv.most_similar(positive=word, topn=n)
```

Getting 3 most similar words of random words along with cosine similarity

```python
for i in range(10):
    word = data[rand.randrange(len(data))]
    try:
        print(word, ':', most_similar(word, 3))
    except KeyError as err:
        print(err)
```
